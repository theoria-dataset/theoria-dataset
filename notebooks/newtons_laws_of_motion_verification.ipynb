## Description
Newton's laws establish the foundation of classical mechanics. The first law defines inertial frames where objects maintain constant velocity absent external forces. The second law relates force to mass times acceleration. The third law requires interaction forces between bodies be equal and opposite, ensuring momentum conservation. These axioms enable prediction of mechanical behavior from projectile motion to orbital dynamics.

## Programmatic Verification

The following code verifies the derivation mathematically:

```python
import sympy as sp

# =====================================================
# Programmatic verification: Newton's Laws of Motion
#
# This mirrors the derivation steps:
#  - Step 1: First law - zero net force implies constant velocity
#  - Step 2: Second law - F = m*a definition
#  - Step 3: Third law - action-reaction pairs
#  - Step 4: First law as special case of second law
#  - Steps 5-6: Conservation of momentum from second and third laws
#
# Note: The entry uses vector notation (bb prefix) but this verification
# uses 1D scalar components for simplicity. The algebraic relationships
# verified here hold component-wise and generalize directly to 3D vectors.
# =====================================================

# Symbols and parameters
t = sp.Symbol('t', real=True)
m, m1, m2 = sp.symbols('m m_1 m_2', positive=True, real=True)

# Vector components for position, velocity, acceleration
# Using scalar components for 1D verification (generalizes to 3D)
r = sp.Function('r')(t)
v = sp.Function('v')(t)
a = sp.Function('a')(t)

# For two-body system
v1 = sp.Function('v_1')(t)
v2 = sp.Function('v_2')(t)

# Force symbols
F = sp.Symbol('F', real=True)
F_12 = sp.Symbol('F_12', real=True)  # Force on object 2 by object 1
F_21 = sp.Symbol('F_21', real=True)  # Force on object 1 by object 2

# ---------------------------
# Step 1: Newton's First Law
# sum F = 0 implies dv/dt = 0 (constant velocity)
# ---------------------------
# When net force is zero, acceleration is zero
# This is a definitional statement about inertial frames
# We verify: if dv/dt = 0, then v = constant

v_const = sp.Symbol('v_0', real=True)  # constant velocity
v_first_law = v_const  # velocity when no force acts
dv_dt_first_law = sp.diff(v_const, t)

assert dv_dt_first_law == 0, "First law: zero force implies zero acceleration"

# ---------------------------
# Step 2: Newton's Second Law
# F = m * a = m * dv/dt = m * d^2r/dt^2
# ---------------------------
# Define acceleration as second derivative of position
a_def = sp.diff(r, t, 2)
v_def = sp.diff(r, t)

# Verify a = dv/dt = d^2r/dt^2
assert sp.diff(v_def, t) == a_def, "Acceleration is second derivative of position"

# Second law: F = m * a
F_second_law = m * a_def

# Verify the form F = m * d^2r/dt^2
assert F_second_law == m * sp.diff(r, t, 2), "Second law: F = m * d^2r/dt^2"

# ---------------------------
# Step 3: Newton's Third Law
# F_12 = -F_21
# ---------------------------
# Action-reaction: forces are equal and opposite
third_law_relation = sp.Eq(F_12, -F_21)

# Verify: F_12 + F_21 = 0
sum_forces = F_12 + F_21
sum_forces_substituted = sum_forces.subs(F_12, -F_21)
assert sum_forces_substituted == 0, "Third law: action-reaction forces sum to zero"

# ---------------------------
# Step 4: First law as special case of second law
# F = 0 implies m*a = 0 implies a = 0 (since m > 0)
# ---------------------------
# When F = 0, from F = m*a we get m*a = 0
# Since m > 0 (positive mass), a must be 0

F_zero = sp.Integer(0)
# From F = m*a with F = 0: m*a = 0
# Solve for a given m > 0
a_symbol = sp.Symbol('a', real=True)
eq_zero_force = sp.Eq(m * a_symbol, F_zero)
a_solution = sp.solve(eq_zero_force, a_symbol)[0]

assert a_solution == 0, "Zero force implies zero acceleration"

# ---------------------------
# Steps 5-6: Conservation of momentum
# From second and third laws for isolated two-body system
# ---------------------------

# Second law for each particle:
# F_21 = m1 * dv1/dt (force on particle 1 by particle 2)
# F_12 = m2 * dv2/dt (force on particle 2 by particle 1)

dv1_dt = sp.diff(v1, t)
dv2_dt = sp.diff(v2, t)

# By second law, express forces in terms of accelerations:
F_21_from_second_law = m1 * dv1_dt
F_12_from_second_law = m2 * dv2_dt

# Define total momentum
p_total = m1 * v1 + m2 * v2

# Time derivative of total momentum
dp_total_dt = sp.diff(p_total, t)

# Expand: dp/dt = m1*dv1/dt + m2*dv2/dt
dp_total_dt_expanded = m1 * dv1_dt + m2 * dv2_dt

# Verify dp/dt = m1*dv1/dt + m2*dv2/dt
assert sp.simplify(dp_total_dt - dp_total_dt_expanded) == 0, "Momentum derivative expansion"

# By second law: dp/dt = m1*dv1/dt + m2*dv2/dt = F_21 + F_12
# This shows dp/dt equals sum of internal forces
dp_dt_as_forces = F_21_from_second_law + F_12_from_second_law
assert sp.simplify(dp_total_dt - dp_dt_as_forces) == 0, "dp/dt = F_21 + F_12 by second law"

# By third law: F_12 = -F_21, so F_21 + F_12 = 0
# Substitute third law into the force sum
force_sum_with_third_law = F_21 + (-F_21)
assert force_sum_with_third_law == 0, "Third law: F_21 + F_12 = F_21 + (-F_21) = 0"

# Therefore: dp/dt = F_21 + F_12 = 0 for isolated system
# This completes the derivation chain: Second law gives dp/dt = F_21 + F_12,
# Third law gives F_21 + F_12 = 0, therefore dp/dt = 0

# ---------------------------
# Final verification: momentum is constant
# ---------------------------
# If dp_total/dt = 0, then p_total = constant
# This is the statement of conservation of momentum

p_constant = sp.Symbol('p_0', real=True)  # constant total momentum
dp_constant_dt = sp.diff(p_constant, t)

assert dp_constant_dt == 0, "Constant momentum has zero time derivative"

print("All Newton's Laws verifications passed successfully!")
print("- First Law: Zero net force implies constant velocity")
print("- Second Law: F = m * a = m * d^2r/dt^2")
print("- Third Law: F_12 = -F_21 (action-reaction)")
print("- First Law is special case of Second Law when F = 0")
print("- Conservation of momentum follows from Second and Third Laws")
print("  (dp/dt = F_21 + F_12 by 2nd law, F_21 + F_12 = 0 by 3rd law)")
```
